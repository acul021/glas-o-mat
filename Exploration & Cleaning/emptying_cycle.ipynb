# Emptying cycle

```python
'''
add locations id
get how many containers are in each location
get a view of when a location was approached
calculate how long the cycle is
check if the cycle is always the same
map different cycles on street view
check if when a location is approached, if there is an entry for every container
'''
```

```python
# Add Location ID
df_ContainerActivities['LOCATION_ID'] = df_ContainerActivities['CONTAINER_ID'].apply(lambda x: str(x)[:-4])
df_ContainerActivities
```

```python
# add latitude and longitude to ContainerActivities by merging with Locations
df_Locations['LOCATION_ID'] = df_Locations['LOCATION_ID'].astype(str)
df_ContainerActivities = pd.merge(df_ContainerActivities, df_Locations[['LOCATION_ID', 'GEO_LAT', 'GEO_LON']], on='LOCATION_ID', how='left')
df_ContainerActivities
```

```python
# drop rows that are placeholders or have inaccurate coordinates
df_placeholder = df_ContainerActivities[
    (df_ContainerActivities['GEO_LAT'] < 47) | (df_ContainerActivities['GEO_LAT'] > 55) | (
                df_ContainerActivities['GEO_LON'] < 6) | (df_ContainerActivities['GEO_LON'] > 15)]
df_ContainerActivities = df_ContainerActivities[~df_ContainerActivities.index.isin(df_placeholder.index)]

df_invalid_coordinates = df_ContainerActivities[(df_ContainerActivities['LATITUDE'] < 47) | (df_ContainerActivities['LATITUDE'] > 55) | (df_ContainerActivities['LONGITUDE'] < 6) | (df_ContainerActivities['LONGITUDE'] > 15)]
df_ContainerActivities = df_ContainerActivities[~df_ContainerActivities.index.isin(df_invalid_coordinates.index)]

#df_ContainerActivities = df_ContainerActivities[df_ContainerActivities['IS_EMPTIED'] == 1]
df_ContainerActivities = df_ContainerActivities.dropna(subset=['GEO_LAT'])

df_ContainerActivities
```

```python
# Get a view of when a location was approached
df_ContainerActivities['RECORDED_AT'] = pd.to_datetime(df_ContainerActivities['RECORDED_AT']).dt.date
visit_summary = df_ContainerActivities.groupby('LOCATION_ID').agg(
    LATITUDE=('LATITUDE', 'first'),
    LONGITUDE=('LONGITUDE', 'first'),
    GEO_LAT=('GEO_LAT', 'first'),
    GEO_LON=('GEO_LON', 'first'),
    VISIT_DATES=('RECORDED_AT', lambda x: sorted(list(x.unique()))),
    VISIT_COUNT=('RECORDED_AT', 'nunique')
)
visit_summary
```

```python
#ToDO: The coordinates are only the first visit, they don't consider moving containers
```

```python
visit_summary.reset_index(inplace=True)
visit_summary
```

```python
def calculate_cycle_and_diffs(dates):
    if len(dates) < 2:
        return None, [], None
    date_diffs = pd.Series(dates).diff().dt.days.dropna()
    avg_cycle = date_diffs.mean()
    diff_list = date_diffs.tolist()
    std_dev = date_diffs.std()
    return avg_cycle, diff_list, std_dev
```

```python
visit_summary[['CYCLE_DAYS', 'DIFFERENCES', 'STD_DEV']] = visit_summary['VISIT_DATES'].apply(
    lambda dates: pd.Series(calculate_cycle_and_diffs(dates))
)
visit_summary
```

## Map different cycles on street view

```python
import osmnx as ox
import matplotlib.pyplot as plt
import folium
```

```python
def categorize_cycle(cycle):
    if cycle <= 3:
        return '1-3 days'
    elif 4 <= cycle <= 7:
        return '4-7 days'
    elif 8 <= cycle <= 14:
        return '8-14 days'
    else:
        return 'Over 14 days'
```

```python
color_map = {
    '1-3 days': 'green',
    '4-7 days': 'yellow',
    '8-14 days': 'orange',
    'Over 14 days': 'red'
}
```

```python
visit_summary['CYCLE_CATEGORY'] = visit_summary['CYCLE_DAYS'].apply(categorize_cycle)
```

## Investigate locations that were approach 1-2 days in a row

```python
# return only the location id and the date of the rows where the time difference is 1 or 2 days
filtered_data = []

for _, row in visit_summary.iterrows():
    location_id = row['LOCATION_ID']
    visit_dates = pd.to_datetime(row['VISIT_DATES'])
    differences = row['DIFFERENCES']
    
    for i, diff in enumerate(differences):
        if diff in [1, 2]:
            filtered_data.append([location_id, visit_dates[i], visit_dates[i + 1]])

df_visited_filtered = pd.DataFrame(filtered_data, columns=['LOCATION_ID', 'DATE_1', 'DATE_2'])
df_visited_filtered['TIME_DIFFERENCE'] = (df_visited_filtered['DATE_2'] - df_visited_filtered['DATE_1']).dt.days
df_visited_filtered
```

```python
# parse the dates to datetime
df_ContainerActivities['RECORDED_AT'] = pd.to_datetime(df_ContainerActivities['RECORDED_AT'])
df_visited_filtered['DATE_1'] = pd.to_datetime(df_visited_filtered['DATE_1'])
df_visited_filtered['DATE_2'] = pd.to_datetime(df_visited_filtered['DATE_2'])
```

```python
# get the container activities for the filtered data
df_visited_filtered_activities = pd.merge(df_visited_filtered, df_ContainerActivities, left_on=['LOCATION_ID', 'DATE_1'], right_on=['LOCATION_ID', 'RECORDED_AT'], how='inner')
df_visited_filtered_activities = pd.merge(df_visited_filtered_activities, df_ContainerActivities, left_on=['LOCATION_ID','CONTAINER_ID' ,'DATE_2'], right_on=['LOCATION_ID', 'CONTAINER_ID' ,'RECORDED_AT'], how='inner')
df_visited_filtered_activities.columns = df_visited_filtered_activities.columns.str.replace('_x', '_1').str.replace('_y', '_2')

df_visited_filtered_activities
```
