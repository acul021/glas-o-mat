```python
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
kmz_file = '/Users/annele/PycharmProjects/Glas-O-Mat/data/modified_data/aggregated_output.csv.kmz'
csv_file = '/Users/annele/PycharmProjects/Glas-O-Mat/data/modified_data/aggregated_output.csv'
```

```python
# extract the kml file from the kmz file
with zipfile.ZipFile(kmz_file, 'r') as z:
    kml_filename = [name for name in z.namelist() if name.endswith('.kml')][0]
    with z.open(kml_filename) as kml_file:
        tree = ET.parse(kml_file)
        root = tree.getroot()
        
namespace = {'kml': 'http://www.opengis.net/kml/2.2'}
```

```python
# extract the data from the kml file
data = []
for placemark in root.findall('.//kml:Placemark', namespace):
    name = placemark.find('kml:name', namespace)
    description = placemark.find('kml:description', namespace)
    coordinates = placemark.find('.//kml:coordinates', namespace)
    
    name = name.text if name is not None else 'Unknown'
    description = description.text if description is not None else 'No description'
    coordinates = coordinates.text.strip() if coordinates is not None else ''

    data.append([name, description, coordinates])
```

```python
# save the data as a csv file
df = pd.DataFrame(data, columns=['Name', 'Description', 'Coordinates'])
df.to_csv(csv_file, index=False)

print(f"The file was saved as {csv_file}.")
```

###### Get dates and materials

```python
from datetime import datetime
import re

def parse_string(string):
    parts = string.split('+')  # Teile den String anhand des '+'
    materials = []
    dates = []
    
    for part in parts:
        material, date_str = part.split('_', 1)
        material = material.strip()

        date_str = re.sub(r'#\d+', '', date_str)

        try:
            date_obj = datetime.strptime(date_str, '%Y-%m-%d')
            dates.append(date_obj)
        except ValueError:
            print(f"Fehler bei der Umwandlung des Datums: {date_str}")

        materials.append(material)
    
    return materials, dates
```

```python
df_maps[['material', 'date_list']] = df_maps['Name'].apply(lambda x: pd.Series(parse_string(x)))
df_maps
```

```python
import folium
from folium.plugins import MarkerCluster
from IPython.display import display
```

```python
# Create a base map centered around the average coordinates
center_lat = df_ContainerActivities['LATITUDE'].mean()
center_lon = df_ContainerActivities['LONGITUDE'].mean()
map_ = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Add a marker cluster to group nearby points
marker_cluster = MarkerCluster().add_to(map_)

# Add container activity markers
for _, activity_row in df_ContainerActivities.iterrows():
    folium.Marker(
        location=[activity_row['LATITUDE'], activity_row['LONGITUDE']],
        popup=(
            f"Location ID: {activity_row['LOCATION_ID']}<br>"
            f"Date: {activity_row['RECORDED_AT']}<br>"
            f"Distance: {activity_row['DISTANCE']} meters<br>"
            f"Container ID: {activity_row['CONTAINER_ID']}"
        ),
        icon=folium.Icon(color='blue')
    ).add_to(marker_cluster)

# Add location center markers (from GEO_LAT, GEO_LONG columns in container_activities.csv)
unique_locations = df_ContainerActivities[['LOCATION_ID', 'GEO_LAT', 'GEO_LON']].drop_duplicates()
for _, loc_row in unique_locations.iterrows():
    folium.Marker(
        location=[loc_row['GEO_LAT'], loc_row['GEO_LON']],
        popup=f"Location Center<br>Location ID: {loc_row['LOCATION_ID']}",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map_)

# Display the map
display(map_)
```

```python
import pandas as pd
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
```

```python
# Example DataFrame (replace this with your data)
# container_activities = pd.read_csv("container_activities.csv")

# Initialize the Dash app
app = Dash(__name__)

# Get a list of unique container IDs for the dropdown
unique_containers = df_ContainerActivities['CONTAINER_ID'].unique()

# Layout of the Dash app
app.layout = html.Div([
    html.H1("Container Movement Visualization", style={'textAlign': 'center'}),
    dcc.Dropdown(
        id='container-dropdown',
        options=[{'label': f'Container {container_id}', 'value': container_id} for container_id in unique_containers],
        placeholder='Select a Container',
        style={'width': '50%', 'margin': '0 auto'}
    ),
    dcc.Graph(id='container-map'),
])

# Callback to update the map based on selected container
@app.callback(
    Output('container-map', 'figure'),
    [Input('container-dropdown', 'value')]
)
def update_map(selected_container):
    if not selected_container:
        return go.Figure()  # Return empty figure if no container is selected
    
    # Filter data for the selected container
    container_data = df_ContainerActivities[df_ContainerActivities['CONTAINER_ID'] == selected_container]
    
    # Create the plot
    fig = go.Figure()
    
    # Add starting point
    fig.add_trace(go.Scattergeo(
        lon=[container_data['GEO_LON'].iloc[0]],
        lat=[container_data['GEO_LAT'].iloc[0]],
        mode='markers',
        marker=dict(size=10, color='red'),
        name='Starting Point'
    ))
    
    # Add activity points
    fig.add_trace(go.Scattergeo(
        lon=container_data['LONGITUDE'],
        lat=container_data['LATITUDE'],
        mode='markers+text',
        marker=dict(size=7, color='blue'),
        text=[f"Date: {row['RECORDED_AT']}<br>Distance: {row['DISTANCE']} m"
              for _, row in container_data.iterrows()],
        name='Activity Points'
    ))
    
    # Add lines from starting point to activities
    for _, row in container_data.iterrows():
        fig.add_trace(go.Scattergeo(
            lon=[row['GEO_LON'], row['LONGITUDE']],
            lat=[row['GEO_LAT'], row['LATITUDE']],
            mode='lines',
            line=dict(color='gray', width=1),
            showlegend=False
        ))
    
    # Configure the layout
    fig.update_layout(
        title=f"Container {selected_container} - Movement Plot",
        geo=dict(
            showland=True,
            landcolor="lightgray",
            showcountries=True,
            countrycolor="white"
        )
    )
    
    return fig

# Run the app
app.run_server(mode="inline", debug=True)
```
