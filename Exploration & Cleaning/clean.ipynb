# Drop Test-Phone

```python
df_ContainerActivities_filtered = df_ContainerActivities[~df_ContainerActivities['PHONE_ID'].str.startswith('TKQ1', na=False)]
df_ContainerActivities_filtered
```

# Drop Darmstadt Container

```python
# add Location ID
df_ContainerActivities_filtered['LOCATION_ID'] = df_ContainerActivities_filtered['CONTAINER_ID'].apply(lambda x: str(x)[:-4])
df_ContainerActivities_filtered
```

```python
# filter Darmstadt
locations_filtered = df_Locations[df_Locations['POSTALCODE'].astype(str).str.startswith(('64', '0'))]
locations_filtered
```

```python
# drop containers in Darmstadt
df_ContainerActivities_filtered['LOCATION_ID'] = df_ContainerActivities_filtered['LOCATION_ID'].astype(str)
locations_filtered['LOCATION_ID'] = locations_filtered['LOCATION_ID'].astype(str)

df_ContainerActivities_filtered = df_ContainerActivities_filtered[~df_ContainerActivities_filtered['LOCATION_ID'].isin(locations_filtered['LOCATION_ID'])]
df_ContainerActivities_filtered
```

```python
# save to csv
df_ContainerActivities_filtered.to_csv(os.path.join(folder_path, 'modified_data/ContainerActivities_filtered.csv'), index=False)
```
