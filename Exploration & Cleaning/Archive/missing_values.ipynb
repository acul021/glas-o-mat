# Missing values

## Numbers of containers per location

```python
# Add Location ID
df_ContainerActivities['LOCATION_ID'] = df_ContainerActivities['CONTAINER_ID'].apply(lambda x: str(x)[:-4])
df_ContainerActivities['RECORDED_AT'] = pd.to_datetime(df_ContainerActivities['RECORDED_AT']).dt.date
df_ContainerActivities
```

```python
# Get how many containers are in each location
container_summary = df_ContainerActivities.groupby('LOCATION_ID').agg(
    CONTAINER_LIST=('CONTAINER_ID', lambda x: list(x.unique())),
    CONTAINER_COUNT=('CONTAINER_ID', 'nunique')
)
container_summary.reset_index(inplace=True)
container_summary
```

## Container Activities per location

```python
# Overview of emptying activities per location
container_activity = df_ContainerActivities.groupby(['LOCATION_ID', 'RECORDED_AT']).agg(
    CONTAINER_LIST_APPROACH=('CONTAINER_ID', lambda x: list(x.unique()))
).reset_index()
container_activity['CONTAINER_COUNT_APPROACH'] = container_activity['CONTAINER_LIST_APPROACH'].apply(len)
container_activity
```

```python
# merge both dataframes on Location ID
container_activity = container_activity.merge(container_summary, on='LOCATION_ID')
container_activity
```

## Missing values

```python
# add two columns, one with the difference between container count and container count approach, second a list of missing containers
container_activity['MISSING_CONTAINER_LIST'] = container_activity.apply(lambda x: list(set(x['CONTAINER_LIST']) - set(x['CONTAINER_LIST_APPROACH'])), axis=1)
container_activity['MISSING_CONTAINER_COUNT'] = container_activity['CONTAINER_COUNT'] - container_activity['CONTAINER_COUNT_APPROACH']
container_activity
```
